```python
DIR_SCRIPT = './scripts/'
movies = open('all_movies.txt','r')
url_base = 'https://imsdb.com/scripts/'
not_found = []

for movie in tqdm(list(movies)):
    
    movie = movie.replace(':','')
    
    url = url_base + movie.replace(' ','-')[:-1]+'.html'

    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    
    if soup.find('pre'):
        html_soup = soup.find('pre')     
    elif soup.find('td',class_='scrtext'):
        html_soup = soup.find('td',class_='scrtext')
    else:
        not_found.append(url)
        continue

    page = open(DIR_SCRIPT + movie.replace(' ','-').replace(':','').replace('?','')[:-1]+'.txt','w', encoding="utf-8")
    page.write(html_soup.text)
    
    page.close()
```

```python
not_found_new = []
for i in not_found:
    not_found_new.append(i.split('/')[-1][:-4]+'txt')
```
